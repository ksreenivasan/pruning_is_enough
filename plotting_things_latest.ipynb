## CIFAR

```python
hc_iter_75 = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_75_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/acc_and_sparsity.csv')
hc_iter_50 = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_50_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/acc_and_sparsity.csv')
#hc_iter_25 = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_25_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/acc_and_sparsity.csv')
imp = pd.read_csv('results/CIFAR10/imp_result.csv') # got from pickle.py in download folder

hc_iter_wt = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_75_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/orig_mask.csv')
hc_iter_shuffle_mask_wt = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_75_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/shuffle_mask.csv')
hc_iter_shuffle_weight_wt = pd.read_csv('results/results_pruning_CIFAR10_cResNet18_hc_iter_75_reg_var_red_2_0_001_adam_cosine_lr_0_05_0_1_50_fan_False_signed_constant_unif_width_1_0_seed_42/shuffle_weights.csv')




```

```python
fig, ax1 = plt.subplots(figsize=(15, 10))
ax2 = ax1.twinx()

ax1.plot(imp['epoch'], imp['test_acc'], linewidth=line_width-2, c='C0', label='IMP (150 epochs, Prune every T=25 epochs)', linestyle='dotted')#, marker='^', markersize=12)
ax2.plot(imp['epoch'], imp['model_sparsity'], linewidth=line_width-2, c='C1', label='Weak LTH (w/ rewinding, Prune every 5 epochs)', linestyle='dotted')#, marker='^', markersize=12)

ax1.plot(hc_iter_75['epoch'], hc_iter_75['test_acc'], linewidth=line_width-2, c='C0', linestyle='solid')#, label='HyperCube (150 epochs, T=75)'#, marker='o', markersize=12) # + Weight Training (50 epochs)
ax2.plot(hc_iter_75['epoch'], hc_iter_75['model_sparsity'], linewidth=line_width-2, c='C1', linestyle='solid')#, label='HyperCube (150 epochs, T=75)'#, marker='o', markersize=12) # + Weight Training (50 epochs)

#ax1.plot(hc_iter_50['epoch'], hc_iter_50['test_acc'], linewidth=line_width-2, c='C0', label='HyperCube (150 epochs, T=50)', linestyle='dashed')#, marker='^', markersize=12) # + Weight Training (50 epochs)
#ax2.plot(hc_iter_50['epoch'], hc_iter_50['model_sparsity'], linewidth=line_width-2, c='C1', label='HyperCube (150 epochs, T=50)', linestyle='dashed')#, marker='^', markersize=12) # + Weight Training (50 epochs)

ax1.plot(hc_iter_wt['epoch']+150, hc_iter_wt['test_acc'], linewidth=line_width-2, c='C0', label='HyperCube (150 epochs, T=75) + weight training', linestyle='solid')#, marker='o', markersize=12) # + Weight Training (50 epochs)

ax1.plot(hc_iter_shuffle_mask_wt['epoch']+150, hc_iter_shuffle_mask_wt['test_acc'], linewidth=line_width-2, c='C0', label='HyperCube (150 epochs, T=75) + mask shuffling + weight training', linestyle='dashed')#, marker='o', markersize=12) # + Weight Training (50 epochs)

ax1.plot(hc_iter_shuffle_weight_wt['epoch']+150, hc_iter_shuffle_weight_wt['test_acc'], linewidth=line_width-2, c='C0', label='HyperCube (150 epochs, T=75) + weight shuffling + weight training', linestyle='dashdot')#, marker='o', markersize=12) # + Weight Training (50 epochs)


plt.title("ResNet-18 on CIFAR-10", fontsize=title_size)
ax1.tick_params(axis='both', labelsize=tick_size)
ax2.tick_params(axis='both', labelsize=tick_size)

ax1.set_xlabel("Epoch ", fontsize=label_size)
ax1.set_ylabel("Test Accuracy", fontsize=label_size, color='C0')
ax2.set_ylabel("Model Sparsity", fontsize=label_size, color='C1')

#ax2.set_ylim([0, 55])
ax1.set_ylim([10, 95])
#ax1.set_xlim([0,152])
ax1.legend(fontsize=legend_size, loc='center left', bbox_to_anchor=(1.1, 0.8))#, loc='right')
ax1.grid(True)
#ax2.grid(True)

plt.savefig('HC variants and imp on cifar10_resnet18.jpg', dpi=400)
#plt.savefig('DG (follow low positive) .jpg', dpi=400)
```
